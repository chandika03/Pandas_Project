# Part 2: Data Filtering & Selection

### List all years where Nigeria had a systemic crisis.

```python
nigeria_crisis_year = df[(df['country'] == 'Nigeria') & (df['systemic_crisis'] == 1)]
```

```python
nigeria_crisis_year
```

```python
nigeria_crisis_year = df[(df['country'] == 'Nigeria') & (df['systemic_crisis'] == 1)]['year'].unique()
```

```python
nigeria_crisis_year
```

### Find all countries that had a banking crisis in the year 2000.

```python
countries_banking_crisis = df[(df['banking_crisis'] == 'crisis') & (df['year'] == 2000)]['country'].unique()
```

```python
countries_banking_crisis
```

### Select only the rows where domestic debt default occurred.

```python
debt_occured = df[df['domestic_debt_in_default'] == 1]
```

```python
debt_occured
```

### Which countries had an inflation crisis during the 1990s?

```python
inflation_crisis = df[(df['inflation_crises'] == 1) & (df['year'] == 1990)]
```

```python
inflation_crisis
```

### What was the exchange rate (exch_usd) for South Africa in 2010?

```python
exchange_rate = df[(df['country'] == 'South Africa') & (df['year'] == 2010)]['exch_usd']
```

```python
exchange_rate
```

# Part 3: Grouping & Aggregation

### For each country, count the total number of systemic crises.

```python
total_number = df.groupby('country')['systemic_crisis'].sum()
```

```python
total_number
```

### Which country had the highest average inflation (inflation_annual_cpi)?

```python
average_inflation = df.groupby('country')['inflation_annual_cpi'].mean()
```

```python
average_inflation
```

```python
highest_average_inflation = average_inflation.idxmax()
```

```python
highest_average_inflation
```

### Group by year and find how many currency crises occurred per year.

```python
currency_crises = df.groupby('year')['currency_crises'].sum()
```

```python
currency_crises
```

### For each country, find the average GDP-weighted default.

```python
average_GDP = df.groupby('country')['gdp_weighted_default'].mean()
```

```python
average_GDP
```

# Part 4: Visualization

### Plot the number of systemic crises over time (year vs number of crises).

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['year'],df['systemic_crisis'], color = 'red')
plt.title("number of systemic crises over time")
plt.xlabel("Year")
plt.ylabel("Number of Systematic Crises")
plt.show()
```

### Plot a bar chart showing the top 5 countries with the most banking crises.

```python
crisis_df = df[df['banking_crisis'] == 'crisis']
```

```python
crisis_df
```

```python
crisis_count = crisis_df['country'].value_counts().nlargest(5)
```

```python
crisis_count
```

```python
plt.title('top 5 countries with the most banking crises')
plt.xlabel("Country")
plt.ylabel("number of crisis")
crisis_count.plot(kind = 'bar', color = "red")
plt.show()
```
