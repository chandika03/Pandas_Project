### Part 2: Data Filtering & Selection

```python
ch_no = df[df['director'] == 'Christopher Nolan']['title']
```

```python
ch_no
```

### Find all TV Shows released in 2020.

```python
all_shows_2020 = df[df['release_year'] == 2020]['title']
```

```python
all_shows_2020
```

### Which Movies have a rating of 'PG-13' and were released after 2015?

```python
movie_2015 = df[
                (df['type'] == 'Movie') & 
                (df['rating'] == 'PG-13') & 
                (df['release_year'] > 2015)]
```

```python
movie_2015
```

```python
movie_2015[['title','release_year','rating']]
```

### Find all shows added to Netflix in January 2020.

```python
df['date_added_modified'] = df['date_added'].str.strip()
```

```python
df['date_added_modified'] = pd.to_datetime(df['date_added_modified'],format='mixed', errors='coerce')
```

```python
movie_find = df[(df['type'] == 'TV Show') & (df['date_added_modified'].dt.year == 2020) & (df['date_added_modified'].dt.month ==1)]
```

```python
movie_find
```

### List all Movies produced in India.

```python
df_india = df[(df['country'] == 'India') & (df['type'] == 'Movie')]['title']
```

```python
df_india
```

## Part 3: Grouping & Aggregation

### Which country has the most number of shows/movies?

```python
count_num = df.groupby('country')['type'].count()
```

```python
count_num.idxmax()
```

```python
count_num.max()
```

### How many Movies vs TV Shows were added each year?

```python
df['year_added'] = df['date_added_modified'].dt.year
```

```python
result = df.groupby(['year_added', 'type']).size().unstack().fillna(0).astype(int)
```
